```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython import display
```

```python
datasets = np.load('train_dataset.npz')
lattice_sets = np.load('train_dataset.npz')[datasets.files[0]]
L = lattice_sets.shape[1]
```

```python
counter = range(L)
x,y,z = np.meshgrid(counter, counter, counter)

count = 0
for lattice in lattice_sets[:50]:
    display.clear_output(wait=True)
    display.display()
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x,y,z, c=lattice.flat,s=50,cmap='coolwarm') # flat return np.flatiter generator, s=size of points
    plt.show()
    print(count)
    count += 1
```

```python
import matplotlib as mpl
mpl.use('Qt4Agg') # switch from non-interactive to interative and vice-versa
from mpl_toolkits.mplot3d import Axes3D
```

```python
L=4
lattice = np.arange(L**3).reshape((L,L,L))
#print(lattice)
lattice[1,2,3]=80
print(lattice)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

counter = range(L)
y,x,z = np.meshgrid(counter, counter, counter)

size = [2*np.e**(a/8) for a in lattice.flat] # let the point size scaled non-linearly
ax.scatter(x,y,z, c=lattice.flat,s=size,cmap=plt.cm.hot,zdir='z') # flat return np.flatiter generator

# generating text labels for points
ls=[]
for i in range(L):
    for j in range(L):
        for k in range(L):
            ls.append((i,j,k))

ans = [(i,j) for i,j in zip(ls,lattice.flat)]
for i,j in zip(ls,lattice.flat):
    ax.text(*i,str(j),fontsize=16)

ax.set_xlabel('x')
ax.set_ylabel('y')
#plt.gca().invert_yaxis() # invert y-axis
ax.set_zlabel('z')

plt.show()
```
